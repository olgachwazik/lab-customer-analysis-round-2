```python
import pandas as pd
data = pd.read_csv("marketing_customer_analysis.csv")
data.shape
```

```python
data
```

```python
def clean_headers ():
    cols = [data.columns[i].lower().replace(' ','_') for i in range (len(data.columns))]
    data.columns = cols
    return display(data)
```

```python
clean_headers()
```

```python
data.isna().sum()
```

```python
def remove_nans():
    data["months_since_last_claim"] = data["months_since_last_claim"].fillna(data["months_since_last_claim"].median())
    data["number_of_open_complaints"] = data["number_of_open_complaints"].fillna(data["number_of_open_complaints"].median())
    data["state"] = data["state"].fillna("Unknown")
    data["response"] = data["response"].fillna("Unknown")
    data["vehicle_class"] = data["vehicle_class"].fillna("Unknown")
    data["vehicle_size"] = data["vehicle_size"].fillna("Unknown")
    data["vehicle_type"] = data["vehicle_type"].fillna("Unknown")
    return display(data)
```

```python
remove_nans()
```

```python
def extract_month():
    data["effective_to_date"] = pd.to_datetime(data["effective_to_date"])
    data["month"] = data["effective_to_date"].dt.month
    return data["month"]
```

```python
extract_month()
```

```python
def filter_quarter ():
    cond = data["month"] <= 3
    return data[cond].sort_values("month")
```

```python
filter_quarter()
```
